```python
class Dog:
    # Constructor
    def __init__(self, name, breed=None):
        self.name = name
        self.breed = breed
    
    def bark(self):
        print("barking")

    def describe(self):
        print(f"{self.name} is a {self.breed}")

    number_of_legs = 4 # Static attritbute/class attribute


dog1 = Dog("Tommy", "Golden Retriever")
dog2 = Dog("Rex")

# Self object inbuilt is passed via objects
dog1.bark()
dog1.describe()
```

```python
dog1.number_of_legs # Accessing class attribute
```

```python
# Inheritence
class Animal:
    def __init__(self, name):
        self.name = name

class Cat(Animal):
    def __init__(self, name, furry):
        super().__init__(name) # Call parent's constructor
        self.furry = furry

class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def bark(self):
        print(f"{self.name} says woof!")


dog1 = Dog("Martha", "Pitbull")
dog1.bark()
```

```python
# Lambda
def square(x):
    return x**2

square_lambda = lambda x:x ** 2

print(f"standard: {square(7)} lambda: {square_lambda(7)}")
```
