## Program 13: Fibonacci (N Terms)

**Problem**: Read `N` and print first `N` Fibonacci numbers.

**Input**:
- One integer `N`

**Output**:
- Print Fibonacci numbers separated by space

**Sample Input**:
```
7
```
**Sample Output**:
```
0 1 1 2 3 5 8
```


````python
n = int(input("Enter integer n:"))
fibPrev = 0
fibCurr = 1
if(n <=0): print("No fibonacci number")
elif(n == 1): print("0")
else:
    print(fibPrev)
    print(fibCurr)
for _ in range(2, n):
    fibPrev, fibCurr = fibCurr, fibCurr + fibPrev
    print(fibCurr)
````

## Program 14: Prime Check

```
29
```



````python
n = int(input("Enter integer n:"))
if n<=1: 
    print("Not Prime")
else:
    for i in range(2, n):
        if(n%i == 0): print("Not Prime")
        break
    else: print("Prime")
````

## Program 16: Sum of Digits

**Problem**: Read an integer `n` and print the sum of its digits.

**Input**:
- One integer `n` (assume non-negative for simplicity)

**Output**:
- Sum of digits

**Sample Input**:
```
1205
```
**Sample Output**:
```
8
```


````python
n = int(input("Enter integer n:"))
sum = 0
while n > 0:
    sum += n%10 # Get last digit - lsb - remainder
    n //=10 # Remove last digit - lsb - quotient
print(sum)
````
